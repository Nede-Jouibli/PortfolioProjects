## Scrapping a product page from Amazon

- We will get the title and price from an Amazon page
- Create a csv file to store the data 
- Automate the storing process 
- Send emails in case of a price drop

```python
#importing libraries

from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib
```

```python
#Connect to Amazon

URL = 'https://www.amazon.com/Nikon-COOLPIX-Digital-Camera-Black/dp/B01C3LEBW6/ref=sr_1_10?crid=1K28XM4LGQDO&keywords=camera&qid=1666844318&qu=eyJxc2MiOiI4LjkxIiwicXNhIjoiOC42OSIsInFzcCI6IjguMTIifQ%3D%3D&sprefix=cam%2Caps%2C236&sr=8-10'

headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}

page = requests.get(URL, headers=headers)

soup1= BeautifulSoup(page.content, 'html.parser')
soup2= BeautifulSoup(soup1.prettify(), 'html.parser')

title = soup2.find(id='productTitle').get_text()
price = soup2.find(id='corePriceDisplay_desktop_feature_div').get_text()

print(title)
print(price)

```

```python
#Cleaning Data
price = price.strip()[1:]
title = title.strip()
```

```python
import datetime

today=datetime.date.today()
print(today)
```

```python
#Create a CSV file to store data
import csv

header = ['Title', 'Price', 'Date']
data = [title, price, today]

#with open('AmazonWebScraper.csv', 'w', newline='', encoding='UTF8') as f:
    #writer = csv.writer(f)
    #writer.writerow(header)
    #writer.writerow(data)
```

```python
#automating the appendingprocess

def check_price():
    URL = 'https://www.amazon.com/Nikon-COOLPIX-Digital-Camera-Black/dp/B01C3LEBW6/ref=sr_1_10?crid=1K28XM4LGQDO&keywords=camera&qid=1666844318&qu=eyJxc2MiOiI4LjkxIiwicXNhIjoiOC42OSIsInFzcCI6IjguMTIifQ%3D%3D&sprefix=cam%2Caps%2C236&sr=8-10'

    headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}

    page = requests.get(URL, headers=headers)

    soup1= BeautifulSoup(page.content, 'html.parser')
    soup2= BeautifulSoup(soup1.prettify(), 'html.parser')

    title = soup2.find(id='productTitle').get_text()
    price = soup2.find(id='corePriceDisplay_desktop_feature_div').get_text()

    price = price.strip()[1:]
    title = title.strip()

    import datetime

    today=datetime.date.today()
    print(today)
    import csv

    header = ['Title', 'Price', 'Date']
    data = [title, price, today]

    with open('AmazonWebScraper.csv', 'w', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerow(data)
    if(price < 100):
        send_mail()
    
```

```python
#Checking the price daily
while(True):
    check_price()
    time.sleep(86400)
```
